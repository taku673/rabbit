## Section 1  -入力層～中間層-

- 機械学習・深層学習の入出力の目的による分類  
 1. 識別：データを目的のクラスに分類する。  
 $p(C_k|x)$を求める。
 2. 生成：特定のクラスのデータを生成する。  
 $p(x|C_k)$を求める。

- 識別器(Classifier)の開発アプローチ  
識別機の開発には3つのアプローチがある。（「パターン認識と機械学習」で採用された分類）  
 1. 生成モデル  
 $p(x|C_k),p(C_k)$を推定したうえで、ベイズの定理から$P(C_k|X)$を求める。  
 （※ 積分の計算量が大変そうだが……？）  
 2. 識別モデル  
 $p(C_k|x)$を推定する。  
 3. 識別関数  
 入力値を直接クラスに写像する関数f(x)を推定する。

- 生成モデルと識別モデルの比較  
データの分布は分類結果よりも複雑なことがある。  
生成モデルはこれを（つまり、複雑なデータのありようの全体を）推定する。  
識別モデルでは、分類ができるレベルで確率の推定をすればよく、データを細かく再現することは求めない。  
  
- 識別モデルと識別関数  
識別モデルは確率を出力する、識別関数は識別結果を直接出力する。


- 万能近似定理  
ニューラルネットワークを使えば、なんでも自由に近似ができるのではないか……？

- ニューラルネットワークの全体像  
ニューラルネットワークは、  
 1. 入力データを直接受け取る、【入力層】  
 2. ネットワークとして途中の処理をする【中間層】  
 3. 確率的に識別結果を出力する【出力層】  
から構築されている。  
学習では、重みパラメータ$\mathbb{w}$とbを決定していく。

- 入力層から中間層  
各ニューロンでは、入力された値に重みwを掛けて足し合わせ、出力に変換していく。  
数式で言うと、  
$$
u = \mathbb{w}^T\cdot\mathbb{x} + b\\
z = f(u)
$$  
の形になる。なお、係数wは傾きを制御し、bは切片（平行移動）を制御している。  

- 例  
動物種類分類のネットワークなら、  
体調、体重、ひげの本数、毛の平均長、耳の大きさ、眉間/目鼻距離比、足の長さ、などのパラメータに重みと切片を与え、次の層へ情報を伝播させることを考える。  
※ 重みwが直接パラメータに係るので、基本的にパラメータの正規化は必要ないと思われるが、極端に差がある場合には、正規化等の処理をしたほうが、学習係数を調整しやすいかもしれない。

```python
def print_vec(text, vec):
    print("*** "+ text + " ***")
    print(vec)
    print("")
```

```python
# 重み係数
W = np.array([[0.1], [0.2]])
print_vec("重み", W)
```

```python
# バイアス
b = 0.5
print_vec("バイアス", b)
```

```python
# 入力値(1データ分のみ),横ベクトル
x = np.array([[2,3]])
print_vec("入力", x)
```

```python
# 総入力
u = np.dot(x,W) + b
print_vec("総入力", u)
```

```python
def relu(x):
    return np.maximum(0,x)
z = relu(u)
print_vec("中間層出力", z)
```

3層ネットワークの順伝播

```python
def init_network() -> dict:
    print("#### ネットワークの初期化 ####")
    network = {}
    
    # 入力xは成分数2で、それぞれのデータは横ベクトル。
    network["W1"] = np.array([[0.1, 0.3, 0.5],
                              [0.2, 0.4, 0.6]])
    network["W2"] = np.array([[0.1, 0.4],
                              [0.2, 0.5],
                              [0.3, 0.6]])
    # 出力は2つ
    network["W3"] = np.array([[0.1, 0.3],
                              [0.2, 0.4]])
    network["b1"] = np.array([0.1, 0.2, 0.3])
    network["b2"] = np.array([0.1, 0.2])
    network["b3"] = np.array([1, 2])
    
    print_vec("重み1", network["W1"])
    print_vec("重み2", network["W2"])
    print_vec("重み3", network["W3"])
    print_vec("バイアス1", network["b1"])
    print_vec("バイアス2", network["b2"])
    print_vec("バイアス3", network["b2"])
    
    return network
```

```python
def forward(network : dict, x : np.ndarray):
    print("##### 順伝播開始 #####")
    
    #活性化関数はrelu
    u1 = np.dot(x, network["W1"]) + network["b1"]
    z1 = relu(u1)
    
    u2 = np.dot(z1, network["W2"]) + network["b2"]
    z2 = relu(u2)
    
    # 出力層は恒等関数で出力を決定する。
    u3 = np.dot(z2, network["W3"]) + network["b3"]
    y = u3
    
    print_vec("総入力1", u1)
    print_vec("中間層出力1", z1)
    print_vec("総入力2", u2)
    print_vec("出力2", z2)
    print("出力合計:" + str(np.sum(z2)))
    
    return y, z1, z2
```

```python
x = np.array([[1., 2. ]])
print_vec("入力", x)
network = init_network()
y, z1, z2 = forward(network, x)
```

## Section 2 活性化関数

- 活性化関数
出力する際、計算された値（u）をそのまま使うのではなく、何らかの関数による変換を行う。  
活性化関数としては、非線形な関数を使う。（恒等関数を使う場合もある）  
- 中間層として用いられる関数の例  
 1. ステップ関数(微分の処理ができないので、使いにくい)  
 2. シグモイド関数($\frac{1}{1+e^{-u}}$)  
 3. RELU関数（ランプ関数）(x(x>0),0(x<=0))となる。
- 出力層で使う活性化関数  
中間層の出力は次の層で使うものだが、出力層の出力は我々が実際に欲しいデータ（例えば識別の確率など）  
 1. ソフトマックス関数  
 2. 恒等関数  
 3. シグモイド関数

以下、活性化関数として用いられる各種関数をコードで実装する。

```python
# ステップ関数
def step(x):
    return np.where(x>0, 1, 0)

# シグモイド関数
def sigmoid(x):
    return 1/(1+np.exp(-x))

# ランプ関数
def relu(x):
    return np.maximum(0,x)
```

## Section 3 出力層

- 出力層の役割  
出力層では実際に我々にとって意味のある数値（クラスに分類される確率）を出力する。  
- 誤差関数  
ニューラルネットワークの学習では、現在のNNと訓練データの誤差を減らすことを考える。
そのため、出力層から出てきた出力と、訓練データの正解を比べる。  
どのくらい合っていたか（＝NNの性能がどれだけ良いか）を誤差関数で表す。
$$
E_n(\mathbb{w}) = \frac{1}{2}||(\mathbb{y}-\mathbb{d})||^2
$$
ここで、yは正解ラベル（事前に、手動でラベル付けしたものを想定）、dはNNからの出力。

- 誤差関数を2乗する理由  
各ラベルごとに、正の誤差を持つ物と負の誤差を持つものがあり、打ち消しあってしまうのを防ぐため。  
絶対値を取るような操作でも、問題はないが、計算のやりやすさから、2乗する操作が採用される（ケースが多い）  
- 1/2する理由  
誤差関数には任意の定数を掛けても結果は変わらない。
通常、誤差関数を利用したネットワークのが学習では、計算の途中で微分を用いるため、予め1/2をつけることで微分後の式を簡単にしている。

誤差関数のコードの例

```python
# 平均二乗誤差
def mean_squared_error(d, y):
    return np.mean(np.square(d - y)) / 2

# クロスエントロピー誤差
def cross_entropy_error(d, y):
    if y.ndim == 1:
        d = d.reshape(1, d.size)
        y = y.reshape(1, y.size)

    if d.size == y.size:
        d = d.argmax(axis=1)

    batch_size = y.shape[0]
    return -np.sum(np.log(y[np.arange(batch_size), d] + 1e-7)) / batch_size
```

- 用いる誤差関数  
分類問題では、cross_entropy_errorを用いる。回帰の問題ではmean_squared_errorを用いる。

- 出力層と中間層の活性化関数の違い  
 - 中間層：閾値の前後で信号の強弱を調整
 - 出力層：信号の大きさ（比率）はそのままに変換  
 　　　　分類問題の場合、出力層の出力は0～1の範囲に限定し、総和を1とする必要がある。  

- 出力層で用いる活性化関数  
 1. 回帰の場合
 回帰では、活性化関数に恒等写像$f(u) = u$を、誤差関数に二乗誤差を用いる。  
 2. 二値分類の場合  
 二値分類では、活性化関数にシグモイド関数$f(u)=\frac{1}{1+e^{-u}}$を、誤差関数に交差エントロピー誤差を用いる。  
 3. 多クラス分類の場合  
 多クラス分類では、活性化関数にソフトマックス関数$f(i,u)=\frac{e^{u_i}}{\sum_{k=1}^Ke^{u_k}}$を、誤差関数に交差エントロピー誤差を用いる。

活性化関数のコード例

```python
def sigmoid(x):
    return 1/(1+np.exp(-x))

def softmax(x : np.ndarray) -> np.ndarray:
    # ifはミニバッチ対応）
    if x.ndim == 2:
        x = x.T
        x = x - np.max(x, axis = 0)
        y = np.exp(x) / np.sum(np.exp(x), axis = 0)
        return y.T
    
    # オーバーフロー対策
    x = x - np.max(x)
    
    # 本質的な計算部分
    return np.exp(x) / np.sum(np.exp(x))

# 平均2乗誤差（meanを取っている）
def mean_squared_error(d, y):
    return np.mean(np.square(d - y)) / 2

# クロスエントロピー誤差
def cross_entropy_error(d, y):
    if y.ndim == 1:
        d = d.reshape(1, d.size)
        y = y.reshape(1, y.size)

    if d.size == y.size:
        d = d.argmax(axis=1)

    batch_size = y.shape[0]
    
    # 本質的な計算部分
    # 1e-7は対数関数の0の極限に進むことを防いでいる。
    return -np.sum(np.log(y[np.arange(batch_size), d] + 1e-7)) / batch_size
```

## Section 4 勾配降下法

- 学習  
ニューラルネットワークでは、パラメータwを適切に調整することで機能する。  
そのパラメータwを実データによって最適化プロセスを学習という。  
過去のパラメータwによる推定の間違い部分を訂正しながら、学習を進めていく。  
- 学習率  
学習では、誤差関数の微分に比例した値だけ重みを更新する。  
比例定数（の逆符号）を学習率として設定する。  
学習率が大きすぎると、発散してしまう。  
学習率が小さすぎると、学習に大きな時間がかかってしまう。  

- 確率的勾配降下法  
学習に使うデータから、ランダムに抽出したサンプルを用いて学習を行う。  
 メリット  
 1. データが冗長な場合に計算コストの軽減  
 2. 極小解に収束するリスクを軽減  
 3. オンライン学習（逐次でデータが追加されていくようなケース）に対応できる。  
 これを用いると、一度に使用するメモリの量も抑制できる。

- ミニバッチ勾配降下法  
バッチ学習は全データを使う。  
ミニバッチ勾配降下法では、例えば「512データずつ」のように少しずつのデータセットを学習に利用する。  
メモリに乗り切らないようなデータでも扱うことができる。  
深層学習で、通常使用される方法。(特に断りなくこの方法が用いられる）  
- メリット  
確率的勾配降下法のメリットを損なわず、計算機の計算資源を有効活用できる。  
小分けにした各バッチは並行して計算することができる（SIMD）。
 
- エポックごとの学習の様子の図示

以下確率的勾配降下法のコード例

```python
def d_mean_squared_error(d,y):
    if type(d) == np.ndarray:
        batch_size = d.shape[0]
        dx = (y-d) / batch_size
    else:
        dx = y-d
    return dx
    
def d_sigmoid(x):
    return (1.0 - sigmoid(x)) * sigmoid(x)
```

```python
# これだと、単一のニューロンでも回帰可能？
def f(x):
    y = 3 * x[0] + 2 * x[1]
    return y

def init_network():
    network = {}
    nodesNum = 10
    network["W1"] = np.random.randn(2, nodesNum)
    network["b1"] = np.random.randn(nodesNum)
    network["W2"] = np.random.randn(nodesNum)
    network["b2"] = np.random.randn()
    
    return network

def forward(network, x):
    W1, b1 = network["W1"], network["b1"]
    W2, b2 = network["W2"], network["b2"]
    
    # 入力層
    # x : 2要素の横ベクトル np.array([[a,b]])の形
    u1 = np.dot(x, W1) + b1
    z1 = relu(u1)
    
    # 出力層
    # 活性化関数は恒等関数
    u2 = np.dot(z1, W2) + b2
    y = u2
    
    return z1, y

def backword(x, d, z1, y, network):
    # 誤差関数をW1,W2,b1,b2のそれぞれで偏微分した値を辞書（grad）に入れて返す。
    grad = {}
    
    W1, b1 = network["W1"], network["b1"]
    W2, b2 = network["W2"], network["b2"]
    
    delta2 = d_mean_squared_error(d,y)
    grad["b2"] = np.sum(delta2, axis = 0)
    grad["W2"] = np.dot(z1.T, delta2)
    
    delta1 = np.dot(delta2, W2.T) * d_sigmoid(z1)
    
    delta1 = delta1[np.newaxis, :]
    grad["b1"] = np.sum(delta1, axis = 0)
    x = x[np.newaxis, :]
    grad["W1"] = np.dot(x.T, delta1)
    
    print_vec("誤差のy微分", delta2)
    print_vec("偏微分_バイアス2", grad["b2"])
    print_vec("偏微分_重み2", grad["W2"])

    print_vec("連鎖率u微分まで", delta1)
    print_vec("偏微分_バイアス1", grad["b1"])
    print_vec("偏微分_重み1", grad["W1"])
    
    return grad
```

```python
data_sets_size = 100000
data_sets = [0 for i in range(data_sets_size)]

for i in range(data_sets_size):
    data_sets[i] = {}
    data_sets[i]["x"] = np.random.rand(2)
    
    data_sets[i]["d"] = f(data_sets[i]["x"])
    
losses = []
learning_rate = 0.07

epoch = 1000

network = init_network()

# ここでランダムにデータを抽出している。
random_datasets = np.random.choice(data_sets, epoch)

for dataset in random_datasets:
    x,d = dataset["x"], dataset["d"]
    z1, y = forward(network, x)
    grad = backword(x,d,z1, y, network)
    
    # 勾配降下
    for key in ("W1", "W2", "b1", "b2"):
        network[key] -= learning_rate * grad[key]
        
    loss = mean_squared_error(d,y)
    losses.append(loss)

print("####結果表示####")
lists = range(epoch)
plt.plot(lists, losses, ".")
plt.show()
```

## Section 5 誤差逆伝播法

- 誤差勾配の計算  
$\nabla E$を計算機上で計算する際には、微小量hに対して、  
$$
\frac{\partial E}{\partial w_m} \approx \frac{E(w_m+h)-E(w_m-h)}{2h}
$$  
の計算をするのが一般的。  
ただし、この計算では、非常に計算量が大きくなってしまう。  
→対策として、誤差逆伝播法を使う。  
- 誤差逆伝播法  
ニューラルネットワークの出力側から、誤差量を計算するが、  
途中の各層で、「出力を入力で微分した値」を掛け合わせていく（微分の連鎖率）ことで、入力層までの微分値を算出できる。  
- 具体例  
仮に、誤差関数に2乗誤差、活性化関数に恒等写像、を用いた出力層の誤差購買を計算すると、求めたいのは、
$$
\frac{\partial E}{\partial w_{ji}^{(l)}} = \frac{\partial E}{\partial y}\frac{\partial y}{\partial u}\frac{\partial u}{\partial w_{ji}^{(l)}}
$$
となる。  
各部分の計算として、
$$
 \frac{\partial E}{\partial y} = y -d\\
 \frac{\partial y}{\partial u} = 1\\
 \frac{\partial u}{\partial w_{ji}^{(l)}} = 
 \left( \begin{array}{c}
 0 \\
 \cdot\cdot\cdot\\
 z_i^{l-1}\\
 \cdot\cdot\cdot\\
 0
 \end{array}
 \right)
$$
となる。  
これを計算すると、  
$$
\begin{eqnarray*}
\frac{\partial E}{\partial w_{ji}^{(l)}} &=& \frac{\partial E}{\partial y}\frac{\partial y}{\partial u}\frac{\partial u}{\partial w_{ji}^{(l)}}\\
&=&(\mathbb{y}-\mathbb{d})\cdot
\left( \begin{array}{c}
0\\
\cdot\cdot\cdot\\
z_i\\
\cdot\cdot\cdot\\
0
\end{array} \right)\\
&=&(y_j-d_j)z_i
\end{eqnarray*}
$$  
が求められる。これによって、出力層のパラメータwの更新ができる。  
通常使われる活性化関数や誤差関数は微分が解析的にできるので、同様の手続きによって、活性化関数のパラメータ微分を求めることができる。

- ディープラーニングの開発環境  
誤差逆伝播法などの学習には、それなりのマシンスペックが必要。  
ローカルPC(CPU/GPU/FPGA/ASIC(TPU)）の場合とクラウド（例えば、AWS,GCPなどのサービスを利用）の場合がある。  
ASICについては、Googleのクラウドサービスで利用する方法がとられることが多い。  
- 過学習  
パラメータ数が大きい、巨大なNNで過学習（あらゆるデータを覚えてしまう）が発生しやすい。表現力が高すぎるから。  
過学習を避けるには、訓練データとテストデータを分け、テストデータでの性能を計ることで、過学習が進んでいないかを確かめることができる。  
- データ集合の拡張  
データ数は足りない（増やしたい）ことが多い。   
特に写真データが対象の場合、データ拡張ができるばあいがある。  
分類問題において、効果が高い。（密度推定ではデータの水増しができない）  
具体的には、オフセット、JPEG圧縮、ぼかし、拡大縮小、回転、剪断、クロップなどの手法がある。  
実際にはこれらを組み合わせて画像をたくさん作る。  
NNでは、中間層にノイズを入れてデータ拡張するのが効果的な場合もある。  


誤差逆伝播の回数を増やして、Section4の内容を試してみる。  
ここでは、入力層（8ニューロン)、中間層を2層(4ニューロンずつ)、出力層(1ニューロン)として試す。

```python
def init_network2():
    network = {}
    network["W1"] = np.random.randn(2, 8)
    network["b1"] = np.random.randn(1, 8)
    network["W2"] = np.random.randn(8, 4)
    network["b2"] = np.random.randn(1, 4)
    network["W3"] = np.random.randn(4, 4)
    network["b3"] = np.random.randn(1, 4)
    network["W4"] = np.random.randn(4, 1)
    network["b4"] = np.random.randn(1, 1)
    
    return network

def forward2(network, x):
    W1, b1 = network["W1"], network["b1"]
    W2, b2 = network["W2"], network["b2"]
    W3, b3 = network["W3"], network["b3"]
    W4, b4 = network["W4"], network["b4"]
    
    # 入力層
    # x : 2要素の横ベクトル np.array([[a,b]])の形
    u1 = np.dot(x, W1) + b1
    z1 = relu(u1)
    
    u2 = np.dot(z1, W2) + b2
    z2 = relu(u2)
    
    u3 = np.dot(z2, W3) + b3
    z3 = relu(u3)
    
    # 出力層
    # 活性化関数は恒等関数
    u4 = np.dot(z3, W4) + b4
    y = u4
    
    return z1, z2, z3, y

def backword3(x, z1, z2, z3, y, d, network):
    # 誤差関数をW1,W2,b1,b2のそれぞれで偏微分した値を辞書（grad）に入れて返す。
    grad = {}
    
    W1, b1 = network["W1"], network["b1"]
    W2, b2 = network["W2"], network["b2"]
    W3, b3 = network["W3"], network["b3"]
    W4, b4 = network["W4"], network["b4"]
    
    # 以下誤差逆伝播をすすめる。
    delta4 = d_mean_squared_error(d,y)
    grad["b4"] = np.sum(delta4, axis = 0)
    grad["W4"] = np.dot(z3.T, delta4)
    
    delta3 = np.dot(delta4, W4.T) * d_sigmoid(z3)    
    grad["b3"] = np.sum(delta3, axis = 0)
    grad["W3"] = np.dot(z2.T, delta3)

    delta2 = np.dot(delta3, W3.T) * d_sigmoid(z2)    
    grad["b2"] = np.sum(delta2, axis = 0)
    grad["W2"] = np.dot(z1.T, delta2)

    delta1 = np.dot(delta2, W2.T) * d_sigmoid(z1)    
    grad["b1"] = np.sum(delta1, axis = 0)
    grad["W1"] = np.dot(x.T, delta1)

    return grad
```

```python
data_sets_size = 100000
data_sets = [0 for i in range(data_sets_size)]

for i in range(data_sets_size):
    data_sets[i] = {}
    data_sets[i]["x"] = np.random.rand(1,2)
    
    data_sets[i]["d"] = f(data_sets[i]["x"][0])
    
losses = []
learning_rate = 0.07

epoch = 1000

network = init_network2()
random_datasets = np.random.choice(data_sets, epoch)

for dataset in random_datasets:
    x,d = dataset["x"], dataset["d"]
    z1, z2, z3, y = forward2(network, x)
    grad = backword3(x,z1,z2,z3, y,d, network)
    
    # 勾配降下
    for key in network:

        network[key] -= learning_rate * grad[key]
        
    loss = mean_squared_error(d,y)
    losses.append(loss)

print("####結果表示####")
lists = range(epoch)
plt.plot(lists, losses, ".")
plt.show()
```
