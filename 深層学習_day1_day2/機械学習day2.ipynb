## Section1 勾配消失問題

- 中間層の数について
AlexNetでは、7層。最近のNNは100以上の層を持つ。  
画像データは入力が非常に多くなり、分析のために層を多くする。  
中間層を増やしていくと問題が発生する。  
- 勾配消失問題  
誤差逆伝播がうまくいかなくなってしまう問題。  
中間層の数が増えると、掛け合わされる微分のセットが非常に多くなり、1未満の数を大量に掛け合わせることになって計算上0に近づいていってしまう。  


- 連鎖率の例題  
$$
\begin{eqnarray*}
\frac{dz}{dx} &=& \frac{\partial z}{\partial t}\frac{\partial t}{\partial x}\\
&=& 2t \times 1\\
&=& 2(x+y)
\end{eqnarray*}
$$

逆伝播のコード例

```python
import numpy as np
import matplotlib.pyplot as plt

def d_mean_squared_error(d,y):
    if type(d) == np.ndarray:
        batch_size = d.shape[0]
        dx = (y-d)/batch_size
    else:
        dx = y-d
    return dx

def sigmoid(x):
    return 1/(1 + np.exp(-x))

def d_sigmoid(x):
    return (1.0 - sigmoid(x))*sigmoid(x)

def backward(x,d,z1,y,network: dict):
    grad = {}
    
    W1, b1 = network["W1"], network["b1"]
    W2, b2 = network["W2"], network["b2"]
    
    delta2 = d_mean_squared_error(d,y)
    grad["b2"] = np.sum(delta2, axis=0)
    grad["W2"] = np.dot(z1.T, delta2)
    
    delta1 = np.dot(delta2, W2.T) * d_sigmoid(z1)
    delta1 = delta1[np.newaxis, :]
    grad["b1"] = np.sum(delta1, axis=0)
    x = x[np.newaxis, :]
    grad["W1"] = np.dot(x.T, delta1)
    
    return grad
```

- 勾配消失への対策  
 1. 活性化関数の選択  
 勾配消失を起こしにくい活性化関数として、ReLU関数が選択されることがある。  
 ニューロンのうち必要のない部分は微分値がゼロになり、スパース化にも寄与する。
 2. パラメータの初期値の工夫  
 普通は乱数を用いるが、どういった乱数を使うかどうかに選択肢がある。  
 例えば、Xavierという方法を使うと、良いパラメータ設定ができる。
 3. バッチ正規化  
- Xavier  
　標準正規分布に従う乱数に対して、前の層のノード数の平方根で除算した値を初期値に用いる。

- 初期値の設定の効果  
シグモイド関数を活性化関数にしている場合は、正規分布（σ=1）で初期化すると、ノードのほとんどは0 or 1を出力するようになる。⇒微分はほぼ0。⇒学習が進まない。  
一方で、小さな値を掛けて、パラメータを初期化すると、ほとんどのノードが0.5を出力するようになり、NNが機能しない。  
適度な小ささにパラメータを調整することができれば、学習がうまくいく⇒ここで使うのが、Xavier。  
ReLUのような場合には、He初期化をすると、うまく初期化できる。

- バッチ正規化  
通常NNの学習は、訓練用のデータを小さなバッチ（ミニバッチ）に分けて、学習を行う。  
この際、ミニバッチの、偏りを抑制することで、ミニバッチごとの学習効率を上げ、過学習も抑制できる。  
計算は以下のように行われる。  
$$
    \mu_t =\frac{1}{N_t}\sum_{i=1}^{N_t}x_{ni}\\
    \sigma_t^2=\frac{1}{N_t}\sum_{i=1}^{N_t}(x_{ni}-\mu_t)^2\\
    \hat{x}_{ni} = \frac{x_{ni}-\mu_t}{\sqrt{\sigma_t^2+\theta}}\\
    y_{ni}=\gamma x_{ni} + \beta
$$
つまり、統計的な正規化処理＋各パラメータの変倍、移動を行う。

以下で、コードによる実装を見る。

```python
import numpy as np
from collections import OrderedDict
# 正規化周りが面倒なので、配布のmnistを使う。
# from sklearn.datasets import load_digits
from data.mnist import load_mnist
import matplotlib.pyplot as plt
from abc import ABC, abstractmethod
```

```python
class ActivationFunctionBase(ABC):
    @abstractmethod
    def forward(self, x):
        pass
    
    @abstractmethod
    def backward(self, dout):
        pass

class Sigmoid(ActivationFunctionBase):
    def __init__(self):
        self.out = None
        
    def __sigmoid(self, x):
        return 1/(1 + np.exp(-x))
        
    def forward(self, x):
        self.out = self.__sigmoid(x)
        return self.out
        
    def backward(self, dout):
        return dout * (1.0 - self.out) * self.out

class Relu(ActivationFunctionBase):
    def __init__(self):
        self.mask = None
        
    def forward(self, x):
        self.mask = (x <= 0)
        out = x.copy()
        out[self.mask] = 0
        return out
        
    def backward(self, dout):
        dout[self.mask] = 0
        return dout

class Affine:
    def __init__(self, W, b):
        self.W = W
        self.b = b
        
        self.x = None
        self.original_x_shape = None
        self.dW = None
        self.db = None
        
    def forward(self, x):
        self.original_x_shape = x.shape
        x = x.reshape(x.shape[0], -1)
        self.x = x
        out = np.dot(self.x, self.W) + self.b
        
        return out
    
    def backward(self, dout):
        dx = np.dot(dout, self.W.T)
        self.dW = np.dot(self.x.T, dout)
        self.db = np.sum(dout, axis = 0)
        dx = dx.reshape(*self.original_x_shape)
        
        return dx

class SoftMaxWithLoss():
    def __init__(self):
        self.loss = None
        self.y = None
        self.d = None
        
    def __softmax(self, x):
        if x.ndim == 2:
            x = x.T
            x = x - np.max(x, axis = 0)
            y = np.exp(x) / np.sum(np.exp(x), axis = 0)
            return y.T
        
    def __cross_entropy_error(self, d, y):
        if y.ndim == 1:
            d = d.reshape(1, d.size)
            y = y.reshape(1, y.size)
            
        if d.size == y.size:
            d = d.argmax(axis = 1)
            
        batch_size = y.shape[0]
        return -np.sum(np.log(y[np.arange(batch_size), d] + 1e-7)) / batch_size
        
    def forward(self, x, d):
        self.d = d
        self.y = self.__softmax(x)
        self.loss = self.__cross_entropy_error(self.d, self.y)
        
        return self.loss
    
    def backward(self, dout=1):
        batch_size = self.d.shape[0]
        if self.d.size == self.y.size:
            dx = (self.y - self.d) / batch_size
        else:
            dx = self.y.copy()
            dx[np.arange(batch_size), self.d] -= 1
            dx = dx / batch_size
            
        return dx
```

```python
class MultiLayerNet:
    '''
    input_size        : 入力層のノード数(例えば、mnistなら、8x8=64)
    hidden_size_list  :隠れ層のノード数のリスト
    output_size       :出力層のノード数
    activation        :活性化関数 ("relu"または"sigmoid")
    weight_init_std   :重みの初期化方法（数値を、あるいは"Xavier","He"）
    '''
    def __init__(self, input_size, hidden_size_list, output_size, activation="relu", weight_init_std="relu"):
        self.input_size = input_size
        self.output_size = output_size
        self.hidden_size_list = hidden_size_list
        self.hidden_layer_num = len(hidden_size_list)
        self.activation = activation
        self.weight_init_std = weight_init_std
        self.params = {}
        
        # 重みの初期化
        # ここで、self.params["W" + str(idx)]が定義されている。 
        self.__init_weight()
        
        # レイヤの生成,sigmoidまたはreluの選択
        activation_layer = {"sigmoid": Sigmoid,"relu": Relu}
        self.layers = OrderedDict()
        for idx in range(1, self.hidden_layer_num + 1):
            self.layers["Affine" + str(idx)] = Affine(self.params["W" + str(idx)], self.params["b" + str(idx)])
            self.layers["Activation_function" + str(idx)] = activation_layer[self.activation]()
        
        idx = self.hidden_layer_num + 1
        self.layers["Affine" + str(idx)] = Affine(self.params["W" + str(idx)], self.params["b" + str(idx)])
        
        self.last_layer = SoftMaxWithLoss()
        
    def __init_weight(self):
        all_size_list = [self.input_size] + self.hidden_size_list + [self.output_size]
        for idx in range(1, len(all_size_list)):
            scale = self.weight_init_std
            if str(self.weight_init_std).lower() in ("relu", "he"):
                scale = np.sqrt(2.0 / all_size_list[idx - 1])
            elif str(self.weight_init_std).lower() in ("sigmoid", "xavier"):
                scale = np.sqrt(1.0 / all_size_list[idx - 1])
                
            self.params["W" + str(idx)] = scale * np.random.randn(all_size_list[idx - 1], all_size_list[idx])
            self.params["b" + str(idx)] = np.zeros(all_size_list[idx])
    
    def predict(self, x):
        # 順伝播の計算
        for layer in self.layers.values():
            x = layer.forward(x)
        
        return x
    
    def loss(self, x, d):
        y = self.predict(x)
        
        weight_decay = 0
        for idx in range(1, self.hidden_layer_num + 2):
            W = self.params["W" + str(idx)]
        
        return self.last_layer.forward(y,d) + weight_decay
    
    def accuracy(self, x, d):
        y = self.predict(x)
        y = np.argmax(y, axis = 1)
        if d.ndim != 1 : d = np.argmax(d, axis = 1)
        
        accuracy = np.sum(y == d) / float(x.shape[0])
        return accuracy
    
    def gradient(self, x, d):
        # forward
        self.loss(x,d)
        
        # backward
        dout = 1
        dout = self.last_layer.backward(dout)
        
        layerBuf = list(self.layers.values())
        layerBuf.reverse()
        
        for layer in layerBuf:
            dout = layer.backward(dout)
            
        grad = {}
        for idx in range(1, self.hidden_layer_num + 2):
            grad["W" + str(idx)] = self.layers["Affine" + str(idx)].dW
            grad["b" + str(idx)] = self.layers["Affine" + str(idx)].db

        return grad
```

```python
class VanishingGradientTest:
    def __init__(self, activation, weight_init_std, x_train, d_train, x_test, d_test, hidden_size_list=[40,20], iters_num = 2000, batch_size = 100, learning_rate = 0.01):
        self.activation = activation
        self.weight_init_std = weight_init_std
        self.hidden_size_list = [40, 20]
        self.x_train = x_train
        self.d_train = d_train
        self.x_test = x_test
        self.d_test = d_test
        self.input_size = x_train.shape[1]
        self.output_size = d_train.shape[1]
        self.network = MultiLayerNet(input_size = self.input_size, 
                                     hidden_size_list = self.hidden_size_list,
                                     output_size = self.output_size,
                                     activation = self.activation, 
                                     weight_init_std = self.weight_init_std)
        self.iters_num = iters_num
        self.batch_size = batch_size
        self.learning_rate = learning_rate
        self.train_size = self.x_train.shape[0] 
        self.plot_interval = 10
        
    def train(self):
        self.train_loss_list = []
        self.accuracies_train = []
        self.accuracies_test = []
        
        for i in range(self.iters_num):
            batch_mask = np.random.choice(self.train_size, self.batch_size)
            x_batch = self.x_train[batch_mask]
            d_batch = self.d_train[batch_mask]
            
            grad = self.network.gradient(x_batch, d_batch)
            
            for key in ("W1", "W2", "W3", "b1", "b2", "b3"):
                self.network.params[key] -= self.learning_rate * grad[key]
            
            loss = self.network.loss(x_batch, d_batch)
            self.train_loss_list.append(loss)
            
            if (i + 1) % self.plot_interval == 0:
                accr_test = self.network.accuracy(self.x_test, self.d_test)
                self.accuracies_test.append(accr_test)
                accr_train = self.network.accuracy(self.x_train, self.d_train)
                self.accuracies_train.append(accr_train)
                
                print("Generation: " + str(i+1) + ". 正答率（トレーニング） = " + str(accr_train))
                print("                : " + str(i+1) + ". 正答率（テスト） = " + str(accr_test))
    
    def plot_results(self):
        lists = range(0, self.iters_num, self.plot_interval)
        plt.plot(lists, self.accuracies_train, label = "training set")
        plt.plot(lists, self.accuracies_test, label = "test set")
        plt.legend(loc = "lower right")
        plt.title("accuracy")
        plt.xlabel("count")
        plt.ylabel("accuracy")
        plt.ylim(0, 1.0)
```

## Section2 学習率最適化手法

- 学習率における工夫  
学習率（パラメータの更新時の更新量に関する比例定数）をどのように決めるかどうかに各種手法があり、そこを工夫することで、効率的に学習を進めることができる。  
- 学習率が適正でない場合について  
学習率が大きすぎると、いつまでも学習がうまくいかずに発散してしまう。  
学習率が小さすぎると、収束までに時間がかかってしまう。また、局所解に陥ってしまう。  
- よく使われる手法  
よく使われる手法に以下のものがある。  
 1. モメンタム  
 2. AdaGrad  
 3. RMSProp  
 4. Adam  
- モメンタム  
誤差をパラメータで微分したものと学習率の積を減算した後、現在の重みに前回の重みを減算した値と慣性の積を加算する。  
モメンタムではパラメータの更新が比較的なめらかな形で進んでいく。  
$$
V_t = \mu V_{t-1} - \epsilon \nabla E\\
w^{t+1} = w^{(t)} + V_t
$$
- AdaGrad  
誤差をパラメータで微分したものと再定義した学習率の積を減算する。  
$$
h_0 = \theta\\
h_t = h_{t-1}+(\nabla E)^2\\
w^{(t+1)} = w^{(t)} -\epsilon \frac{1}{\sqrt{h_t}+\theta}\nabla E
$$  
勾配の緩やかな斜面に対して最適解を見つけやすいが、学習率が徐々に小さくなるために、鞍点問題を引き起こすことがある。  
- RMSProp  
誤差をパラメータで微分したものと再定義した学習率の積を減算する  
$$
h_t = \alpha h_{t-1} + (1-\alpha )(\nabla E)^2\\
w^{(t+1)} = w^{(t)}-\epsilon \frac{1}{\sqrt{h_t}+\theta}\nabla E
$$  
大域的最適解を見つけやすい。ハイパーパラメータの調整が必要な場合が少ない。  
- Adam  
通常、もっとも良い方法。モメンタムとRMSPropを両方取り入れたような方法。  

以下pythonでの実装を確認する

```python
class OptimizerTest:
    def __init__(self, activation, weight_init_std, x_train, d_train, x_test, d_test, OptimizerType, momentum = 0.9,
                 decay_rate = 0.99, beta1 = 0.9, beta2 = 0.99, hidden_size_list=[40,20], iters_num = 2000, batch_size = 100, learning_rate = 0.01):
        """
        OptimizerTypeは、SGD、Momentum、AdaGrad、RMSprop、ADAのいずれかを文字列で受け取る。
        """
        self.activation = activation
        self.weight_init_std = weight_init_std
        self.hidden_size_list = [40, 20]
        self.x_train = x_train
        self.d_train = d_train
        self.x_test = x_test
        self.d_test = d_test
        self.OptimizerType = OptimizerType
        self.learning_rate = learning_rate
        # 各Optimizer固有の値
        self.momentum = momentum
        self.decay_rate = 0.99
        self.beta1 = beta1
        self.beta2 = beta2
        
        self.input_size = x_train.shape[1]
        self.output_size = d_train.shape[1]
        self.network = MultiLayerNet(input_size = self.input_size, 
                                     hidden_size_list = self.hidden_size_list,
                                     output_size = self.output_size,
                                     activation = self.activation, 
                                     weight_init_std = self.weight_init_std)
        self.iters_num = iters_num
        self.batch_size = batch_size
        self.train_size = self.x_train.shape[0] 
        self.plot_interval = 10
        
    def train(self):
        self.train_loss_list = []
        self.accuracies_train = []
        self.accuracies_test = []
        
        for i in range(self.iters_num):
            batch_mask = np.random.choice(self.train_size, self.batch_size)
            x_batch = self.x_train[batch_mask]
            d_batch = self.d_train[batch_mask]
            
            grad = self.network.gradient(x_batch, d_batch)
            if self.OptimizerType == "Momentum":
                if i == 0:
                    v = {}
            elif self.OptimizerType == "AdaGrad" or self.OptimizerType == "RMSprop":
                if i == 0:
                    h = {}
            elif self.OptimizerType == "Adam":
                if i == 0:
                    m = {}
                    v = {}
                learning_rate_t = self.learning_rate * np.sqrt(1.0 - self.beta2 ** (i + 1)) / (1.0 - self.beta1 ** (i + 1))
            
            for key in ("W1", "W2", "W3", "b1", "b2", "b3"):
                if self.OptimizerType == "SGD":
                    self.network.params[key] -= self.learning_rate * grad[key]
                elif self.OptimizerType == "Momentum":
                    if i == 0:
                        v[key] = np.zeros_like(self.network.params[key])
                    v[key] = self.momentum * v[key] - self.learning_rate * grad[key]
                    self.network.params[key] += v[key]
                elif self.OptimizerType == "AdaGrad":
                    if i == 0:
                        h[key] = np.full_like(self.network.params[key], 1e-4)
                    else:
                        h[key] += np.square(grad[key])
                    self.network.params[key] -= self.learning_rate * grad[key] / (np.sqrt(h[key]))
                elif self.OptimizerType == "RMSprop":
                    if i == 0:
                        h[key] = np.zeros_like(self.network.params[key])
                    h[key] *= self.decay_rate
                    h[key] += (1 - self.decay_rate) * np.square(grad[key])
                    self.network.params[key] -= self.learning_rate * grad[key] / (np.sqrt(h[key]) + 1e-7)
                elif self.OptimizerType == "Adam":
                    if i == 0:
                        m[key] = np.zeros_like(self.network.params[key])
                        v[key] = np.zeros_like(self.network.params[key])
                    m[key] += (1 - self.beta1) * (grad[key] - m[key])
                    v[key] += (1 - self.beta2) * (grad[key] ** 2 -v[key])
                    self.network.params[key] -= learning_rate_t * m[key] / (np.sqrt(v[key]) + 1e-7)
                
            
            loss = self.network.loss(x_batch, d_batch)
            self.train_loss_list.append(loss)
            
            if (i + 1) % self.plot_interval == 0:
                accr_test = self.network.accuracy(self.x_test, self.d_test)
                self.accuracies_test.append(accr_test)
                accr_train = self.network.accuracy(self.x_train, self.d_train)
                self.accuracies_train.append(accr_train)
                
                print("Generation: " + str(i+1) + ". 正答率（トレーニング） = " + str(accr_train))
                print("                : " + str(i+1) + ". 正答率（テスト） = " + str(accr_test))
    
    def plot_results(self):
        lists = range(0, self.iters_num, self.plot_interval)
        plt.plot(lists, self.accuracies_train, label = "training set")
        plt.plot(lists, self.accuracies_test, label = "test set")
        plt.legend(loc = "lower right")
        plt.title("accuracy")
        plt.xlabel("count")
        plt.ylabel("accuracy")
        plt.ylim(0, 1.0)
```

## Section3 過学習

- 過学習とは  
学習用（訓練用）データでは良い出力が出るが、検証用データでは良い出力が出ない状態になってしまう。  
特定の訓練サンプルに特化した学習になってしまっている。  
学習結果をエポック数対誤差でプロットすることで、テスト誤差と訓練誤差の乖離から過学習が進んでいるか判断できる。  
- 過学習の原因  
パラメータ数が多い。パラメータの値が適切でない。ノードが多い。など。ネットワークの自由度が高い（高すぎる）と過学習が起きやすい。  
- 過学習の対策：正則化  
正則化とは、ネットワークの自由度を制限する。  
過学習が起きていると、特定のパラメータが極端に大きくなってしまっている。  
つまり、パラメータが極端な値を取れないようにする。  
⇒誤差に対して、正則化項を加算することで、重みを抑制する。  
正則化項は  
$$
    \frac{1}{p}\lambda ||x||_p\\
    ||x||_p = (|x_1|^p+\cdot\cdot\cdot +|x_n|^p)^{\frac{1}{p}}
$$
と書ける。p=1がL1正則化（マンハッタン距離：各パラメータの和）p=2がL2正則化（ユークリッド距離：二乗和の平方根）である。  
L1正則化では、パラメータが0になりやすく、スパース化が進みやすい。  

- 過学習の対策：ドロップアウト  
ランダムにノードを削除して学習させること。  
データのバリエーションが生まれてデータ拡張の効果が表れる。  

以下、過学習になるようにデータ等を調整して、過学習発生をpython上で確認する。

```python
class Dropout:
    def __init__(self, dropout_ratio = 0.5):
        self.dropout_ratio = dropout_ratio
        self.mask = None
    
    def forward(self, x, train_flg = True):
        if train_flg:
            self.mask = np.random.rand(*x.shape) > self.dropout_ratio
            return x * self.mask
        else:
            return x * (1.0 - self.dropout_ratio)
        
    def backward(self, dout):
        return dout * self.mask
    
```

```python
class MultiLayerNet2:
    '''
    input_size        : 入力層のノード数(例えば、mnistなら、8x8=64)
    hidden_size_list  :隠れ層のノード数のリスト
    output_size       :出力層のノード数
    activation        :活性化関数 ("relu"または"sigmoid")
    weight_init_std   :重みの初期化方法（数値を、あるいは"Xavier","He"）
    use_dropout       :ドロップアウトの有無
    dropout_ratio     :ドロップアウト率
    '''
    def __init__(self, input_size, hidden_size_list, output_size, activation="relu", weight_init_std="relu",
                 use_dropout = False, dropout_ratio = 0.5):
        self.input_size = input_size
        self.output_size = output_size
        self.hidden_size_list = hidden_size_list
        self.hidden_layer_num = len(hidden_size_list)
        self.activation = activation
        self.weight_init_std = weight_init_std
        self.params = {}
        self.use_dropout = use_dropout
        self.dropout_ratio = dropout_ratio
        
        # 重みの初期化
        # ここで、self.params["W" + str(idx)]が定義されている。 
        self.__init_weight()
        
        # レイヤの生成,sigmoidまたはreluの選択
        activation_layer = {"sigmoid": Sigmoid,"relu": Relu}
        self.layers = OrderedDict()
        for idx in range(1, self.hidden_layer_num + 1):
            self.layers["Affine" + str(idx)] = Affine(self.params["W" + str(idx)], self.params["b" + str(idx)])
            self.layers["Activation_function" + str(idx)] = activation_layer[self.activation]()
            
            if self.use_dropout:
                self.layers["Dropout" + str(idx)] = Dropout(self.dropout_ratio)
        
        idx = self.hidden_layer_num + 1
        self.layers["Affine" + str(idx)] = Affine(self.params["W" + str(idx)], self.params["b" + str(idx)])
        
        self.last_layer = SoftMaxWithLoss()
        
    def __init_weight(self):
        all_size_list = [self.input_size] + self.hidden_size_list + [self.output_size]
        for idx in range(1, len(all_size_list)):
            scale = self.weight_init_std
            if str(self.weight_init_std).lower() in ("relu", "he"):
                scale = np.sqrt(2.0 / all_size_list[idx - 1])
            elif str(self.weight_init_std).lower() in ("sigmoid", "xavier"):
                scale = np.sqrt(1.0 / all_size_list[idx - 1])
                
            self.params["W" + str(idx)] = scale * np.random.randn(all_size_list[idx - 1], all_size_list[idx])
            self.params["b" + str(idx)] = np.zeros(all_size_list[idx])
    
    def predict(self, x, train_flg = False):
        # 順伝播の計算
        for key, layer in self.layers.items():
            if "Dropout" in key:
                x = layer.forward(x, train_flg)
            else:
                x = layer.forward(x)
        
        return x
    
    def loss(self, x, d, train_flg = False):
        y = self.predict(x, train_flg)
        
        for idx in range(1, self.hidden_layer_num + 2):
            W = self.params["W" + str(idx)]
        
        return self.last_layer.forward(y,d)
    
    def accuracy(self, x, d):
        y = self.predict(x)
        y = np.argmax(y, axis = 1)
        if d.ndim != 1 : d = np.argmax(d, axis = 1)
        
        accuracy = np.sum(y == d) / float(x.shape[0])
        return accuracy
    
    def gradient(self, x, d):
        # forward
        self.loss(x,d, train_flg=True)
        
        # backward
        dout = 1
        dout = self.last_layer.backward(dout)
        
        layerBuf = list(self.layers.values())
        layerBuf.reverse()
        
        for layer in layerBuf:
            dout = layer.backward(dout)
            
        grad = {}
        for idx in range(1, self.hidden_layer_num + 2):
            grad["W" + str(idx)] = self.layers["Affine" + str(idx)].dW
            grad["b" + str(idx)] = self.layers["Affine" + str(idx)].db

        return grad
```

```python
class OverfittingTest:
    def __init__(self, activation, weight_init_std, x_train, d_train, x_test, d_test, 
                 OptimizerType = "SDG", momentum = 0.9, decay_rate = 0.99, beta1 = 0.9, beta2 = 0.99, 
                 weight_decay_type = None, weight_decay_lambda = 0.01, dropout = False, dropout_ratio = 0.5, 
                 hidden_size_list=[40,20], iters_num = 2000, batch_size = 100, learning_rate = 0.01):
        """
        OptimizerTypeは、SGD、Momentum、AdaGrad、RMSprop、ADAのいずれかを文字列で受け取る。
        weight_dcay_typeは、L1,L2のいずれかを文字列で受け取る。
        """
        self.activation = activation
        self.weight_init_std = weight_init_std
        self.hidden_size_list = hidden_size_list
        self.hidden_layer_num = len(hidden_size_list)
        self.x_train = x_train
        self.d_train = d_train
        self.x_test = x_test
        self.d_test = d_test
        self.OptimizerType = OptimizerType
        self.learning_rate = learning_rate
        # 各Optimizer固有の値
        self.momentum = momentum
        self.decay_rate = 0.99
        self.beta1 = beta1
        self.beta2 = beta2
        # 過学習防止処理
        self.weight_decay_type = weight_decay_type
        self.weight_decay_lambda = weight_decay_lambda
        self.dropout = dropout
        self.dropout_ratio = dropout_ratio
        
        self.input_size = x_train.shape[1]
        self.output_size = d_train.shape[1]
        self.network = MultiLayerNet2(input_size = self.input_size, 
                                     hidden_size_list = self.hidden_size_list,
                                     output_size = self.output_size,
                                     activation = self.activation, 
                                     weight_init_std = self.weight_init_std,
                                     use_dropout = self.dropout,
                                     dropout_ratio = self.dropout_ratio)

        
        self.iters_num = iters_num
        self.batch_size = batch_size
        self.train_size = self.x_train.shape[0] 
        self.plot_interval = 10
        
    def train(self):
        self.train_loss_list = []
        self.accuracies_train = []
        self.accuracies_test = []
        
        for i in range(self.iters_num):
            batch_mask = np.random.choice(self.train_size, self.batch_size)
            x_batch = self.x_train[batch_mask]
            d_batch = self.d_train[batch_mask]
            
            grad = self.network.gradient(x_batch, d_batch)
            if self.OptimizerType == "Momentum":
                if i == 0:
                    v = {}
            elif self.OptimizerType == "AdaGrad" or self.OptimizerType == "RMSprop":
                if i == 0:
                    h = {}
            elif self.OptimizerType == "Adam":
                if i == 0:
                    m = {}
                    v = {}
                learning_rate_t = self.learning_rate * np.sqrt(1.0 - self.beta2 ** (i + 1)) / (1.0 - self.beta1 ** (i + 1))
            
            weight_decay = 0
            for idx in range(1, self.hidden_layer_num + 1):
                if self.OptimizerType == "SGD":
                    if self.weight_decay_type == "L1":
                        grad["W" + str(idx)] = self.network.layers["Affine" + str(idx)].dW + self.weight_decay_lambda * np.sign(self.network.params["W" + str(idx)])
                        grad["b" + str(idx)] = self.network.layers["Affine" + str(idx)].db
                        self.network.params["W" + str(idx)] -= self.learning_rate * grad["W" + str(idx)]
                        self.network.params["b" + str(idx)] -= self.learning_rate * grad["b" + str(idx)]
                        weight_decay += self.weight_decay_lambda * np.sum(np.abs(self.network.params["W" + str(idx)]))                        
                    elif self.weight_decay_type == "L2":
                        grad["W" + str(idx)] = self.network.layers["Affine" + str(idx)].dW + self.weight_decay_lambda * self.network.params["W" + str(idx)]
                        grad["b" + str(idx)] = self.network.layers["Affine" + str(idx)].db
                        self.network.params["W" + str(idx)] -= self.learning_rate * grad["W" + str(idx)]
                        self.network.params["b" + str(idx)] -= self.learning_rate * grad["b" + str(idx)]
                        weight_decay += 0.5 * self.weight_decay_lambda * np.sqrt(np.sum(self.network.params["W" + str(idx)]**2))
                    else:
                        grad["W" + str(idx)] = self.network.layers["Affine" + str(idx)].dW
                        grad["b" + str(idx)] = self.network.layers["Affine" + str(idx)].db
                        self.network.params["W" + str(idx)] -= self.learning_rate * grad["W" + str(idx)]
                        self.network.params["b" + str(idx)] -= self.learning_rate * grad["b" + str(idx)]
                elif self.OptimizerType == "Momentum":
                    if i == 0:
                        v["W" + str(idx)] = np.zeros_like(self.network.params["W" + str(idx)])
                        v["b" + str(idx)] = np.zeros_like(self.network.params["b" + str(idx)])
                    v["W" + str(idx)] = self.momentum * v["W" + str(idx)] - self.learning_rate * grad["W" + str(idx)]
                    v["b" + str(idx)] = self.momentum * v["b" + str(idx)] - self.learning_rate * grad["b" + str(idx)]
                    self.network.params["W" + str(idx)] += v["W" + str(idx)]
                    self.network.params["b" + str(idx)] += v["b" + str(idx)]
                elif self.OptimizerType == "AdaGrad":
                    if i == 0:
                        h["W" + str(idx)] = np.full_like(self.network.params["W" + str(idx)], 1e-4)
                        h["b" + str(idx)] = np.full_like(self.network.params["b" + str(idx)], 1e-4)
                    else:
                        h["W" + str(idx)] += np.square(grad["W" + str(idx)])
                        h["b" + str(idx)] += np.square(grad["b" + str(idx)])
                    self.network.params["W" + str(idx)] -= self.learning_rate * grad["W" + str(idx)] / (np.sqrt(h["W" + str(idx)]))
                    self.network.params["b" + str(idx)] -= self.learning_rate * grad["b" + str(idx)] / (np.sqrt(h["b" + str(idx)]))
                elif self.OptimizerType == "RMSprop":
                    if i == 0:
                        h["W" + str(idx)] = np.zeros_like(self.network.params["W" + str(idx)])
                        h["b" + str(idx)] = np.zeros_like(self.network.params["b" + str(idx)])
                    h["W" + str(idx)] *= self.decay_rate
                    h["b" + str(idx)] *= self.decay_rate
                    h["W" + str(idx)] += (1 - self.decay_rate) * np.square(grad["W" + str(idx)])
                    h["b" + str(idx)] += (1 - self.decay_rate) * np.square(grad["b" + str(idx)])
                    self.network.params["W" + str(idx)] -= self.learning_rate * grad["W" + str(idx)] / (np.sqrt(h["W" + str(idx)]) + 1e-7)
                    self.network.params["b" + str(idx)] -= self.learning_rate * grad["b" + str(idx)] / (np.sqrt(h["b" + str(idx)]) + 1e-7)
                elif self.OptimizerType == "Adam":
                    if i == 0:
                        m["W" + str(idx)] = np.zeros_like(self.network.params["W" + str(idx)])
                        m["b" + str(idx)] = np.zeros_like(self.network.params["b" + str(idx)])
                        v["W" + str(idx)] = np.zeros_like(self.network.params["W" + str(idx)])
                        v["b" + str(idx)] = np.zeros_like(self.network.params["b" + str(idx)])
                    m["W" + str(idx)] += (1 - self.beta1) * (grad["W" + str(idx)] - m["W" + str(idx)])
                    m["b" + str(idx)] += (1 - self.beta1) * (grad["b" + str(idx)] - m["b" + str(idx)])
                    v["W" + str(idx)] += (1 - self.beta2) * (grad["W" + str(idx)] ** 2 -v["W" + str(idx)])
                    v["b" + str(idx)] += (1 - self.beta2) * (grad["b" + str(idx)] ** 2 -v["b" + str(idx)])
                    self.network.params["W" + str(idx)] -= learning_rate_t * m["W" + str(idx)] / (np.sqrt(v["W" + str(idx)]) + 1e-7)
                    self.network.params["b" + str(idx)] -= learning_rate_t * m["b" + str(idx)] / (np.sqrt(v["b" + str(idx)]) + 1e-7)
                
            
            loss = self.network.loss(x_batch, d_batch) + weight_decay
            self.train_loss_list.append(loss)
            
            if (i + 1) % self.plot_interval == 0:
                accr_test = self.network.accuracy(self.x_test, self.d_test)
                self.accuracies_test.append(accr_test)
                accr_train = self.network.accuracy(self.x_train, self.d_train)
                self.accuracies_train.append(accr_train)
                
                print("Generation: " + str(i+1) + ". 正答率（トレーニング） = " + str(accr_train))
                print("                : " + str(i+1) + ". 正答率（テスト） = " + str(accr_test))
    
    def plot_results(self):
        lists = range(0, self.iters_num, self.plot_interval)
        plt.plot(lists, self.accuracies_train, label = "training set")
        plt.plot(lists, self.accuracies_test, label = "test set")
        plt.legend(loc = "lower right")
        plt.title("accuracy")
        plt.xlabel("count")
        plt.ylabel("accuracy")
        plt.ylim(0, 1.0)
```

## Section4 畳み込みニューラルネットワークの概念

- 畳み込みニューラルネットワーク（CNN）とは？  
画像の分析で使うことが多い。  
次元のつながりのデータを効率的に扱える。（つまり、各入力パラメータ同士に順番や並び方に意味がある場合）  
- 画像のCNNのケース  
入力層の後で、畳み込み層/プーリング層をかさね、出力層の直前で全結合層を挟んで出力層が得られる。  
- LuNetのケース  
入力層(32,32)を一層目（畳み込み層）で(28,28,6）の出力に変換される。（畳み込み数では、「評価を6人で行う」のように数を増やす）  
次にプーリング層では要約がなされ、(14,14,6）の出力に変換される。  
応用に畳み込み層で（10,10,16）に変換され、次にプーリング層で(5,5,16）に変換される。  
ここから全結合層でひとまとめにされて、(120,)→(84,)→(10,)と変換され、欲しい出力を得る。  
- 畳み込み層  
入力に対してフィルターを複数用意する（評価を複数人で行うことに対応）。  
畳み込み層では次元的なつながりがあるものは何でも学習できる。  
フィルターの持つ値がニューロンの重みパラメータに対応する。  
フィルターが$n \times n$の時、画像の縦横の画素数を$n-1$だけ減らすことになってしまうため、画像の外側に特定の値があるとみなして、畳み込み処理をすることがある。（パディング）  
また、フィルターを1マスずつ動かす代わりに、1マス飛ばしのようにすることもできる（ストライド）。これを行うと画像サイズが一気に小さくなる。1/ストライド、になる。  
畳み込み演算をする際は、行列の積で演算ができるように、変換する処理が行われる。
- プーリング層  
プーリング層では、畳み込み同様に少しずつずらして値を取得するが、「要約」するような処理が行われる。  
Max Poolingでは、最大値を、Avg. Poolingでは平均値を出力する。
- 全結合層  
いわゆる普通のNNと変わらない。  
CNNでは、畳み込み層/プーリング層では次元のつながりが保たれているが、  
全結合層ではその繋がりをばらして、学習が行われる。

```python
# im2col関数  
def im2col(input_data, filter_h, filter_w, stride=1, pad=0):
    N, C, H, W = input_data.shape
    out_h = (H + 2 * pad - filter_h) // stride + 1
    out_w = (W + 2 * pad - filter_w) // stride + 1
    
    img = np.pad(input_data, [(0,0), (0,0), (pad,pad), (pad,pad)], "constant")
    col = np.zeros((N, C, filter_h, filter_w, out_h, out_w))
    
    for y in range(filter_h):
        y_max = y + stride * out_h
        for x in range(filter_w):
            x_max = x + stride * out_w
            col[:, :, y, x, :, :] = img[:, :, y: y_max:stride, x:x_max:stride]
    col = col.transpose(0,4,5,1,2,3)
    col = col.reshape(N * out_h * out_w, -1)
    return col
```

```python
def col2im(col, input_shape, filter_h, filter_w, stride = 1, pad = 0):
    N, C, H, W = input_shape
    out_h = (H + 2 * pad - filter_h) // stride + 1
    out_w = (W + 2 * pad - filter_w) // stride + 1
    col = col.reshape(N, out_h, out_w, C, filter_h, filter_w).transpose(0, 3, 4, 5, 1, 2)
    
    img = np.zeros((N, C, H + 2 * pad + stride - 1, W + 2 * pad + stride - 1))
    for y in range(filter_h):
        y_max = y + stride * out_h
        for x in range(filter_w):
            x_max = x + stride * out_w
            img[:, :, y:y_max:stride, x:x_max:stride] += col[:, :, y, x, :, :]
            
    return img[:, :, pad:H + pad, pad:W + pad]
```

```python
class Convolution:
    def __init__(self, W, b, stride=1, pad=0):
        self.W = W
        self.b = b
        self.stride = stride
        self.pad = pad
        
        self.x = None
        self.col = None
        self.col_W = None
        
        self.dW = None
        self.db = None
    
    def forward(self, x):
        FN, C, FH, FW = self.W.shape
        N, C, H, W = x.shape
        out_h = 1 + int((H + 2 * self.pad -FH) / self.stride)
        out_w = 1 + int((W + 2 * self.pad -FH) / self.stride)
        
        col = im2col(x, FH, FW, self.stride, self.pad)
        col_W = self.W.reshape(FN, -1).T
        out = np.dot(col, col_W) + self.b
        out = out.reshape(N, out_h, out_w, -1).transpose(0,3,1,2)
        
        self.x = x
        self.col = col
        self.col_W = col_W
        
        return out
    
    def backward(self, dout):
        FN, C, FH, FW = self.W.shape
        dout = dout.transpose(0,2,3,1).reshape(-1, FN)
        
        self.db = np.sum(dout, axis = 0)
        self.dW = np.dot(self.col.T, dout)
        self.dW = self.dW.transpose(1,0).reshape(FN, C, FH, FW)
        dcol = np.dot(dout, self.col_W.T)
        dx = col2im(dcol, self.x.shape, FH, FW, self.stride, self.pad)
        
        return dx
```

```python
class Pooling:
    def __init__(self, pool_h, pool_w, stride = 1, pad = 0):
        self.pool_h = pool_h
        self.pool_w = pool_w
        self.stride = stride
        self.pad = pad
        
        self.x = None
        self.arg_max = None
        
    def forward(self, x):
        N, C, H, W = x.shape
        out_h = int(1 + (H - self.pool_h) / self.stride)
        out_w = int(1 + (W - self.pool_w) / self.stride)
        
        col = im2col(x, self.pool_h, self.pool_w, self.stride, self.pad)
        col = col.reshape(-1, self.pool_h * self.pool_w)
        
        arg_max = np.argmax(col, axis=1)
        out = np.max(col, axis=1)
        out = out.reshape(N, out_h, out_w, C).transpose(0,3,1,2)
        
        self.x = x
        self.arg_max = arg_max
        
        return out
    
    def backward(self, dout):
        dout = dout.transpose(0,2,3,1)
        
        pool_size = self.pool_h * self.pool_w
        dmax = np.zeros((dout.size, pool_size))
        dmax[np.arange(self.arg_max.size), self.arg_max.flatten()] = dout.flatten()
        dmax = dmax.reshape(dout.shape + (pool_size,))
        
        dcol = dmax.reshape(dmax.shape[0] * dmax.shape[1] * dmax.shape[2], -1)
        dx = col2im(dcol, self.x.shape, self.pool_h, self.pool_w, self.stride, self.pad)
        
        return dx
```

```python
class Relu:
    def __init__(self):
        self.mask = None
        
    def forward(self, x):
        self.mask = ( x <= 0)
        out = x.copy()
        out[self.mask] = 0
        
        return out
    
    def backward(self, dout):
        dout[self.mask] = 0
        dx = dout
        
        return dx
```

```python
class SimpleConvNet:
    def __init__(self, input_dim=(1,28,28), conv_param={'filter_num':30, 'filter_size':5, "pad":0, "stride":1},
                 hidden_size = 100, output_size = 10, weight_init_std=0.01):
        filter_num = conv_param["filter_num"]
        filter_size = conv_param["filter_size"]
        filter_pad = conv_param["pad"]
        filter_stride = conv_param["stride"]
        input_size = input_dim[1]
        conv_output_size = (input_size - filter_size + 2 * filter_pad) / filter_stride + 1
        pool_output_size = int(filter_num * (conv_output_size / 2) * (conv_output_size / 2))
        
        # 重みの初期化
        self.params = {}
        self.params["W1"] = weight_init_std * np.random.randn(filter_num, input_dim[0], filter_size, filter_size)
        self.params["b1"] = np.zeros(filter_num)
        self.params["W2"] = weight_init_std * np.random.randn(pool_output_size, hidden_size)
        self.params["b2"] = np.zeros(hidden_size)
        self.params["W3"] = weight_init_std * np.random.randn(hidden_size, output_size)
        self.params["b3"] = np.zeros(output_size)
        
        # レイヤの生成
        self.layers = OrderedDict()
        self.layers["Conv1"] = Convolution(self.params["W1"], self.params["b1"], conv_param["stride"], conv_param["pad"])
        self.layers["Relu1"] = Relu()
        self.layers["Pool1"] = Pooling(pool_h=2, pool_w=2, stride=2)
        self.layers["Affine1"] = Affine(self.params["W2"], self.params["b2"])
        self.layers["Relu2"] = Relu()
        self.layers["Affine2"] = Affine(self.params["W3"], self.params["b3"])
        
        self.last_layer = SoftMaxWithLoss()
        
    def predict(self, x):
        for key in self.layers.keys():
            x = self.layers[key].forward(x)
        return x

    def loss(self, x, d):
        y = self.predict(x)
        return self.last_layer.forward(y,d)
    
    def accuracy(self, x, d, batch_size = 100):
        if d.ndim != 1 : d = np.argmax(d, axis=1)
            
        acc = 0.0
        
        for i in range(int(x.shape[0] / batch_size)):
            tx = x[i*batch_size:(i+1)*batch_size]
            td = d[i*batch_size:(i+1)*batch_size]
            y = self.predict(tx)
            y = np.argmax(y, axis=1)
            acc += np.sum(y == td)
        
        return acc / x.shape[0]
    
    def gradient(self, x, d):
        self.loss(x,d)
        
        dout = 1
        dout = self.last_layer.backward(dout)
        layers = list(self.layers.values())
        
        layers.reverse()
        for layer in layers:
            dout = layer.backward(dout)
            
        grad = {}
        grad["W1"], grad["b1"] = self.layers["Conv1"].dW, self.layers["Conv1"].db
        grad["W2"], grad["b2"] = self.layers["Affine1"].dW, self.layers["Affine1"].db
        grad["W3"], grad["b3"] = self.layers["Affine2"].dW, self.layers["Affine2"].db
        
        return grad
```

```python
class Adam:
    def __init__(self, learning_rate = 0.001, beta1 = 0.9, beta2 = 0.999):
        self.learning_rate = learning_rate
        self.beta1 = beta1
        self.beta2 = beta2
        self.iter = 0
        self.m = None
        self.v = None
    
    def update(self, params, grad):
        if self.m is None:
            self.m, self.v = {},{}
            for key, val in params.items():
                self.m[key] = np.zeros_like(val)
                self.v[key] = np.zeros_like(val)
        
        self.iter += 1
        lr_t = self.learning_rate * np.sqrt(1.0 - self.beta2 ** self.iter) / (1.0 - self.beta1 ** self.iter)
        
        for key in params.keys():
            self.m[key] += (1 - self.beta1) * (grad[key] - self.m[key])
            self.v[key] += (1 - self.beta2) * (grad[key] ** 2 - self.v[key])
            
            params[key] -= lr_t * self.m[key] / (np.sqrt(self.v[key]) + 1e-7)
```

```python
import keras
from keras.models import Sequential
from keras.layers import Dense, Dropout, Activation, Flatten
from keras.layers import Conv2D, MaxPooling2D, BatchNormalization
from keras.initializers import TruncatedNormal, Constant
from keras.utils.vis_utils import plot_model
import matplotlib.pyplot as plt
```

```python
def conv2d(filters, kernel_size, strides=(1,1), padding = "same", bias_init=1, **kwargs):
    trunc = TruncatedNormal(mean=0.0, stddev=0.01)
    cnst = Constant(value=bias_init)
    return Conv2D(
        filters, kernel_size, strides=strides, padding=padding,
        activation="relu", kernel_initializer=trunc, bias_initializer=cnst, **kwargs
    )

def dense(units, activation = "tanh"):
    trunc = TruncatedNormal(mean=0.0, stddev=0.01)
    cnst = Constant(value=1)
    return Dense(
        units, activation=activation,
        kernel_initializer=trunc, bias_initializer=cnst
    )

def AlexNet(image_size, channel, num_classes):
    model=Sequential()
    
    model.add(conv2d(96, 3, bias_init=0, input_shape=(image_size, image_size, channel)))
    model.add(MaxPooling2D(pool_size=(2,2), strides=(2,2)))
    model.add(BatchNormalization())
    
    model.add(conv2d(256,5))
    model.add(MaxPooling2D(pool_size=(2,2), strides=(2,2)))
    model.add(BatchNormalization())
    
    model.add(conv2d(384,3,bias_init=0))
    model.add(conv2d(384,3))
    model.add(conv2d(256,3))
    model.add(MaxPooling2D(pool_size=(2,2), strides=(2,2)))
    model.add(BatchNormalization())
    
    model.add(Flatten())
    model.add(dense(4096))
    model.add(Dropout(0.5))
    
    model.add(dense(4096))
    model.add(Dropout(0.5))
    
    model.add(dense(num_classes, activation="softmax"))
    
    return model
```

```python
from keras.datasets import cifar10
from keras.preprocessing.image import ImageDataGenerator
from keras.optimizers import SGD
import keras.utils
```

```python
batch_size = 128
num_classes = 10
epochs = 100
image_size = 32
channel = 3

(x_train, y_train), ( x_test, y_test) = cifar10.load_data()
y_train = keras.utils.np_utils.to_categorical(y_train, num_classes)
y_test = keras.utils.np_utils.to_categorical(y_test, num_classes)

model = AlexNet(image_size, channel, num_classes)
model.compile(optimizer=SGD(lr=0.01), loss="categorical_crossentropy", metrics=["accuracy"])

model.summary()
```

```python
train_gen = ImageDataGenerator(rescale = 1.0 / 255, horizontal_flip = True,
                              width_shift_range=4.0/32.0, height_shift_range=4.0/32.0)
test_gen = ImageDataGenerator(rescale=1.0/255)
result = model.fit_generator(train_gen.flow(x_train, y_train, batch_size, shuffle=True),
                    steps_per_epoch = x_train.shape[0]//batch_size,
                    validation_data=test_gen.flow(x_test, y_test, batch_size, shuffle=False),
                    validation_steps = x_test.shape[0]//batch_size,
                    max_queue_size=5, epochs=epochs)
```

```python
train_acc = model.history.history["accuracy"]
test_acc = model.history.history["val_accuracy"]
epochs = range(len(train_acc))
plt.plot(epochs,train_acc,label="train accuracy")
plt.plot(epochs, test_acc,label = "test accuracy")
plt.legend()
plt.show()
```
