# 機械学習  
## 主成分分析

- 主成分分析  
次元圧縮を行う。  
情報をできるだけ落とさないように、次元を落とす。  
うまく3次元に落とせれば、人間の目で分かりやすく特徴をつかめる（かもしれない）  
現在の基底に対して、新たな基底を抽出する。抽出時は、
その基底にそって計算されるデータの分散が最大になるような基底を取る。  
「分散が大きい」→「データの情報が多くのこっている」  
新たな基底$\mathbb{a}$のなす座標系でのデータ（ただし、平均を原点に移動）は、  
$$
\mathbb{s}_j = \bar{X}\mathbb{a}_j
$$
と表せる。  
変換後の分散は  
$$
Var(\mathbb{s}_j) = \mathbb{a}_j^TVar(\bar{X})\mathbb{a}_j
$$  
であり、これを最大化するような$\mathbb{a}$を求める。  
ただし、あるaに対して平行なものは解になってしまうため、ノルムが1になる制約を加える。  

- ラグランジュの未定乗数法  
条件付き最大化問題では、ラグランジュ関数を作って、微分 = 0を解けばよい。  
$$
\frac{\partial E(\mathbb{a}_j}{\partial \mathbb{a}_j} = 2Var(\bar{X})\mathbb{a}_j-2\lambda \mathbb{a}_j = 0\\
Var(\bar{X})\mathbb{a}_j = \lambda \mathbb{a}_j
$$  
つまり、$Var(\bar{X})$の固有値、固有ベクトルを求めるということ。  
※ 固有ベクトル自体は無数にあるが、ノルムが1の制約があるため、無数にはならない。  
ここで、$\mathbb{a}$への射影先でのデータの分散は、この固有値と一致する。  
$$
Var(\mathbb{s}_i) = \lambda _i
$$  
最大の$\lambda _i$に対応する$mathbb{a}_i$が、分散を最大化させる第1軸であり、欲しい軸の個数に合わせて、上から$\lambda _i$を選び、対応する$\mathbb{a}_i$を軸とすればよい。

- 寄与率  
行った主成分分析で、どれだけ情報が失われたか。  
第k主成分の寄与率は、  
$$
c_k = \frac{\lambda _k}{\sum_{i=1}^m\lambda_i}
$$  
となり、第k主成分までの累積寄与率は  
$$
r_k = \frac{\sum_{j=1}^k\lambda_j}{\sum_{i=1}^m\lambda_i}
$$  
となる。

以下、sklearnを用いた実践

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegressionCV
from sklearn.metrics import confusion_matrix
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
```

```python
cancer_df = pd.read_csv('./cancer.csv')
```

```python
cancer_df.shape
```

```python
cancer_df.drop('Unnamed: 32',axis=1, inplace=True)
```

```python
# 目的変数の抽出 
y = cancer_df.diagnosis.apply(lambda d: 1 if d == 'M' else 0)
```

```python
X = cancer_df.loc[:,'radius_mean':]
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=0)

scaler=StandardScaler()
X_train_scaled =scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

logistic = LogisticRegressionCV(cv=10, random_state=0)
logistic.fit(X_train_scaled, y_train)

print(f"訓練：{logistic.score(X_train_scaled, y_train)}")
print(f"検証：{logistic.score(X_test_scaled, y_test)}")
print(f'Confusion matrix: \n{confusion_matrix(y_true=y_test, y_pred=logistic.predict(X_test_scaled))}')
```

```python
pca = PCA(n_components=30)
pca.fit(X_train_scaled)
plt.bar([n for n in range(1, len(pca.explained_variance_ratio_)+1)], pca.explained_variance_ratio_)
```

```python
pca = PCA(n_components=2)
X_train_pca = pca.fit_transform(X_train_scaled)
print(f"寄与率：{pca.explained_variance_ratio_}")

temp = pd.DataFrame(X_train_pca)
temp['Outcome'] = y_train.values
b = temp[temp['Outcome'] == 0]
m = temp[temp['Outcome'] == 1]
plt.scatter(x=b[0], y=b[1], marker='o')
plt.scatter(x=m[0], y=m[1], marker='^')
plt.xlabel('PC 1')
plt.ylabel("PC 2")
```

ちなみに、3つまでみると

```python
pca = PCA(n_components=3)
X_train_pca = pca.fit_transform(X_train_scaled)
print(f"寄与率：{pca.explained_variance_ratio_}")

temp = pd.DataFrame(X_train_pca)
temp['Outcome'] = y_train.values
b = temp[temp['Outcome'] == 0]
m = temp[temp['Outcome'] == 1]
fig, ax = plt.subplots(1,3,figsize = (18,6))
ax[0].scatter(x=b[0], y=b[1], marker='o')
ax[0].scatter(x=m[0], y=m[1], marker='^')
ax[1].scatter(x=b[1], y=b[2], marker='o')
ax[1].scatter(x=m[1], y=m[2], marker='^')
ax[2].scatter(x=b[2], y=b[0], marker='o')
ax[2].scatter(x=m[2], y=m[0], marker='^')
plt.show()
```

```python
from mpl_toolkits import mplot3d
fig = plt.figure(figsize = (12,12))
ax = plt.axes(projection="3d")
ax.scatter(b[0],b[1],b[2])
ax.scatter(m[0],m[1],m[2])
plt.show()
```
