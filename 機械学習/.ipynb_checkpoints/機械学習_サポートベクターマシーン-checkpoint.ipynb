# 機械学習
## サポートベクターマシン

- サポートベクターマシン（カーネル導入前の話）  
2クラス分類のためのアルゴリズム。  
$y = \mathbb{w}^T\mathbb{x}+b=\sum_{j=1}^mw_jx_j+b$として、yの符号で分類する。  
この際、wとbを決定するためのアルゴリズムが、サポートベクターマシン。

- wの決め方  
線形判別関数(n-1次元の超平面になる)と最も近いデータ点との距離をマージンという。  
各点から線形判別関数までの距離は$|\mathbb{w}^T\mathbb{x}+b|/||\mathbb{w}||$である。  
サポートベクターマシンでは、このマージンが最大となるようにwとbを決定する。  
数式で言うと、
$$
\max_{w,b}\left[\min_i\frac{t_i(\mathbb{w}^T\mathbb{x}_i+b)}{||\mathbb{w}||}\right]
$$
マージン上の点において、$t_i(\mathbb{w}^T\mathbb{x}+b)=1$が成り立つため、すべての点に対して$t_i(\mathbb{w}^T\mathbb{x}+b)\geq 1$。

- 解くべき条件付き最適化問題  
SVMでの$\mathbb{w}$の決定では、
$$
\min_{w,b}\frac{1}{2}||\mathbb{w}||^2\\
t_i(\mathbb{w}^T\mathbb{x}_i+b)\geq1\qquad (i=1, 2, ...,n)
$$
を解く。通常、ラグランジュの未定乗数法を用いて解く。

- ラグランジュの未定乗数法による解法  
ラグランジュ関数は  
$$
L(\mathbb{w},b,\mathbb{a}) =\frac{1}{2}||\mathbb{w}||^2-\sum_{i=1}^na_i(t_i(\mathbb{w}^T\mathbb{x}_i+b)-1)
$$  
であり、w,bは以下の関係を満たす。
$$
\frac{\partial L}{\partial\mathbb{w}} = \mathbb{w}-\sum_{i=1}^na_it_i\mathbb{x}_i=0\\
\frac{\partial L}{\partial b}=-\sum_{i=1}^na_it_i=0
$$
つまり、
$$
\mathbb{w}=\sum_{i=1}^na_it_i\mathbb{x}_i\\
\sum_{i=0}^na_it_i=0
$$  
を満たす。

- 双対問題  
ラグランジュの未定乗数法の問題は、以下のように書き換えることができる。  
ラグランジュ関数からw,bを消去すると、
$$
\tilde{L}=\sum_{i=1}^na_i-\frac{1}{2}\sum_{i=1}^n\sum_{j=1}^na_ia_jt_it_j\mathbb{x}_i^T\mathbb{x}_j
$$  
aを変化させたときこの関数の最大値が、w,bの関数と見たときの最小値と対応する。  
そこで、双対問題として、
$$
\max_a\sum_{i=1}^na_i-\frac{1}{2}\sum_{i=1}^n\sum_{j=1}^na_ia_jt_it_j\mathbb{x}_i^T\mathbb{x}_j\\
\sum_{i=1}^na_it_i=0\\
a_i\geq0
$$  
を解くことを考える。(この形だと、xの内積の形が現れているのが計算上都合がよい）

- ソフトマージンSVM  
サンプルを線形分離できない場合は、誤差を許容し、誤差にペナルティを与える。  
これによって、線形分離できない場合にも対応できる。また誤差の許容具合はハイパーパラメータCによって制御する。（Cが大きいほど誤差を許容しない）

- 非線形分離  
線形分離できない場合、特徴空間に写像し、その空間で線形分離する。  
これは、線形回帰モデルを非線形回帰モデルに発展させるときに、$\phi (x)$を用いて拡張したのと同様。  
このとき、「単回帰分析」や「重回帰分析」として実施していた部分が、svmになる。  
- カーネルトリック  
解くべき問題の$\mathbb{x}$を$\phi (\mathbb{x})$に置き換えると、
$$
\max_a\sum_{i=1}^na_i^\frac{1}{2}\sum_{i=1}^n\sum_{j=1}^na_ia_jt_it_j\phi (\mathbb{x}_i)^T\phi (\mathbb{x}_j)
$$
となる。通常$\phi (\mathbb{x})$として用いる関数に制限はないが、ある種の性質を持つ関数を使うと、  
$$
k(\mathbb{x}_i,\mathbb{x}_j)=\phi (\mathbb{x}_i)^T\phi (\mathbb{x}_j)
$$
を用いて、計算を簡単にできる。  
実務上、頻繁に用いられるのは、放射基底関数カーネル（ガウシアンカーネル）である。

### 実装  
以下SVMの実装を行う。

```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

訓練データ生成器

```python
class data_generator():
    def __init__(self, sample_type, size):
        self.size = size
        self.sample_type = sample_type
    
    def gen_data(self):
        if self.sample_type == "linear":
            # 乱数を使っているので、低い確率で線形分離不可能になる場合もある。
            x0 = np.random.normal(size = self.size).reshape(-1,2) -2.
            x1 = np.random.normal(size = self.size).reshape(-1,2) +2.
            self.x_train = np.concatenate([x0,x1])
            self.ys_train = np.concatenate([np.zeros(math.floor(self.size/2)), np.ones(math.ceil(self.size/2))]).astype(np.int)
            return self.x_train, self.ys_train
        elif self.sample_type == "circle":
            factor = .2
            linspace = np.linspace(0,2*np.pi, self.size // 2 + 1)[:-1]
            outer_circ_x = np.cos(linspace)
            outer_circ_y = np.sin(linspace)
            inner_circ_x = outer_circ_x * factor
            inner_circ_y = outer_circ_y * factor
            
            self.x_train = np.vstack((np.append(outer_circ_x, inner_circ_x),np.append(outer_circ_y, inner_circ_y))).T
            self.ys_train = np.hstack([np.zeros(self.size // 2, dtype=np.intp),
                           np.ones(self.size // 2, dtype=np.intp)])
            self.x_train += np.random.normal(scale=0.15, size =self.x_train.shape)
            return self.x_train, self.ys_train
        elif self.sample_type == "softmargin":
            # 乱数を使っているので、低い確率で線形分離不可能になる場合もある。
            x0 = np.random.normal(size = self.size).reshape(-1,2) -1.
            x1 = np.random.normal(size = self.size).reshape(-1,2) +1.
            self.x_train = np.concatenate([x0,x1])
            self.ys_train = np.concatenate([np.zeros(math.floor(self.size/2)), np.ones(math.ceil(self.size/2))]).astype(np.int)
            return self.x_train, self.ys_train
        else:
            print("sample_typeの設定が不適切")
            return
        
    def data_plot(self):
        plt.scatter(self.x_train[:,0], self.x_train[:,1], c=self.ys_train)
```

svm

```python
class svm():
    def __init__(self, kernel, eta1 = 0.01, eta2 = 0.001, sigma= 0.8, n_iter = 500, softmargin = False, C=1):
        self.kernel = kernel
        self.eta1 = eta1
        self.eta2 = eta2
        self.sigma = sigma
        self.n_iter = n_iter
        self.softmargin = softmargin
        self.C = C
    
    def fit(self, X_train, ys_train):
        self.X_train = X_train
        self.ys_train = ys_train
        self.n_samples = len(X_train)
        if self.kernel == "linear":
            self.K = self.X_train.dot(self.X_train.T)
        elif self.kernel == "rbf":
            self.K = np.zeros((self.n_samples, self.n_samples))
            for i in range(self.n_samples):
                for j in range(self.n_samples):
                    self.K[i,j] = np.exp(-0.5*((self.X_train[i] - self.X_train[j])**2).sum()/self.sigma**2)
        else:
            print("カーネルが不適切")
        
        self.t = np.where(self.ys_train == 1.0, 1.0 , -1.0)
        self.H = np.outer(self.t,self.t) * self.K
        self.a = np.ones(self.n_samples)
        for _ in range(self.n_iter):
            grad = 1 - self.H.dot(self.a)
            self.a += self.eta1 * grad
            self.a -= self.eta2 * self.a.dot(self.t)*self.t
            if self.softmargin:
                self.a = np.clip(self.a, 0, self.C)
            else:
                self.a = np.where(self.a > 0, self.a, 0)
        self.support_vector_index = self.a > 1e-4
        self.support_vectors = self.X_train[self.support_vector_index]
        self.support_vector_t = self.t[self.support_vector_index]
        self.support_vector_a = self.a[self.support_vector_index]
        term2 = self.K[self.support_vector_index][:,self.support_vector_index].dot(self.support_vector_a * self.support_vector_t)
        self.b = (self.support_vector_t - term2).mean()
        
    def predict(self, X_test):
        y_project = np.ones(len(X_test)) * self.b
        for i in range(len(X_test)):
            for a, sv_t, sv in zip(self.support_vector_a, self.support_vector_t, self.support_vectors):
                if self.kernel == "linear":
                    y_project[i] += a * sv_t * sv.dot(X_test[i])
                elif self.kernel == "rbf":
                    y_project[i] += a * sv_t * np.exp(-0.5*((X_test[i] - sv)**2).sum()/self.sigma**2)                    
        return y_project, np.sign(y_project)
        
        
    def plt_results(self):
        plot_num = 100
        min_x0 = self.X_train[:,0].min()
        max_x0 = self.X_train[:,0].max()
        width_x0 = max_x0 - min_x0
        min_x1 = self.X_train[:,1].min()
        max_x1 = self.X_train[:,1].max()
        width_x1 = max_x1 - min_x1        
        xx0, xx1 = np.meshgrid(np.linspace(min_x0 - width_x0 / 20 , max_x0 + width_x0 / 20, plot_num),np.linspace(min_x1 - width_x1 / 20 , max_x1 + width_x1 / 20, plot_num))
        xx = np.array([xx0, xx1]).reshape(2, -1).T
        
        y_project, y_pred = self.predict(xx)
        plt.scatter(self.X_train[:, 0], self.X_train[:,1], c=self.ys_train)
        plt.scatter(self.support_vectors[:, 0], self.support_vectors[:,1], s=100, facecolors="none", edgecolors='k')
        plt.contour(xx0, xx1, y_project.reshape(plot_num,plot_num), colors="k", levels = [-1,0,1], alpha = 0.5, linestyles=["--","-","--"])
```

```python
DG1 = data_generator("linear", 50)
X_train1, ys_train1 = DG1.gen_data()
DG1.data_plot()
plt.show()
```

```python
SVM1 = svm(kernel = "linear", eta1 = 0.01, eta2 = 0.001, n_iter = 500)
SVM1.fit(X_train1, ys_train1)
SVM1.plt_results()
plt.show()
```

```python
DG2 = data_generator("circle", 50)
X_train2, ys_train2 = DG2.gen_data()
DG2.data_plot()
plt.show()
```

```python
SVM2 = svm(kernel = "rbf", eta1 = 0.01, eta2 = 0.001, sigma = 0.8, n_iter = 5000)
SVM2.fit(X_train2, ys_train2)
SVM2.plt_results()
plt.show()
```

```python
DG3 = data_generator("softmargin", 50)
X_train3, ys_train3 = DG3.gen_data()
DG3.data_plot()
plt.show()
```

```python
SVM3 = svm(kernel = "linear", eta1 = 0.01, eta2 = 0.001, n_iter = 500, softmargin = True, C = 1)
SVM3.fit(X_train3, ys_train3)
SVM3.plt_results()
plt.show()
```

上記プログラムのように、Kの計算の部分のみ書き換えれば、容易に非線形回帰（kernel="rbf"）に応用することができる。
